```python
import numpy as np


#항등함수
def identity(x):
    return x
```

```python
x = np. arange(1,11)
y = identity(x)
x == y
```

```python
def step(x):
    return np. where(x>0,1,0)
```

```python
x = np. arange(-5,5,0.1)
y = step(x)
y
```

```python
import matplotlib.pylab as plt
%matplotlib inline
```

```python
plt.plot(x,y)
plt.grid()
plt.show()
```

```python
#시그모이드함수(sigmoid function)

def sigmoid(x):
    return 1 / (1 + np. exp(-x))


x = np.arange(-5,5,0.1)
y = sigmoid(x)
x.shape, y.shape
```

```python
plt.plot(x,y)
plt.grid()
plt.show()
```
